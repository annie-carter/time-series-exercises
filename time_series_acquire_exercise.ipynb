## STARWARS PEOPLE

### 1. Using the code from the lesson as a guide and the REST API from https://swapi.dev/ as we did in the lesson, create a dataframe named people that has all of the data for people.

```python
#Worked to integrated the code steps for one clean code will use this version for acquire.py 
def get_people_data():
    # Initialize the URL for the first page
    data = []
    url = 'https://swapi.dev/api/people/'  
    
    while True:
        # Send a GET request to the SWAPI
        response = requests.get(url)  
        if response.status_code == 200:
            json_data = response.json()
            results = json_data['results']
            data.extend(results)
            # Update the URL with the next page URL
            url = json_data['next']  

            if url is None:
                break
        else:
            print(f"Error: {response.status_code}")
            break
    # Create a DataFrame using the collected data
    df = pd.DataFrame(data)  
    return df

```
